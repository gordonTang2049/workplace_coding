```python
import pandas as pd
import numpy as np
```

```python
def inactive_cus(df):
    if "inactive" in df["Customer Name"]:
        return "inactive"
    elif "TEMP CLOSED COVID-19" in df["Customer Name"]:
        return "tem_close"
    else:
        return "active"
    
def join_addr(df):
    addr = ''
    if df.unknown is not None:
        addr_1 = [df.unknown, df.addr_1]
        addr = ' - '.join(addr_1)
        return addr 
    else:
        return df.addr_1

def reconstruct_addr(df):
    addr = df['Property Name'].split('-')
    if len(addr) < 3:
        state = ""
        suburb = ""
        addr_1 = ""
        return [state, suburb, addr_1]
    elif len(addr) > 3:
        state = addr[0]
        suburb = addr[1]
        slice_elem = slice(3,4)
        
        addr_1= " - ".join(addr[slice_elem]) 
        print(addr_1)
        return [state, suburb, addr_1]
    else:
        state = addr[0]
        suburb = addr[1]
        addr_1 = addr[2]
        return [state, suburb, addr_1]
```
